```python
import pandas as pd
from math import sqrt

def gde(file_path):
    """
    Processa a planilha Excel para calcular o somatório de D para cada elemento,
    incluindo d_max, gde e gdf para cada elemento em cada planilha.

    Parâmetros:
        file_path (str): Caminho para o arquivo Excel.

    Retorno:
        pd.DataFrame: DataFrame contendo os elementos, o somatório de D (sum(D)),
                      o d_max, o gde e o gdf para cada elemento, incluindo o nome da planilha.
    """
    # Carregar o arquivo Excel com múltiplas planilhas
    excel_data = pd.read_excel(file_path, sheet_name=None, header=[0, 1])  # Carrega todas as planilhas

    # Lista para armazenar os resultados
    resultados = []

    # Iterar sobre cada planilha no arquivo
    for sheet_name, df in excel_data.items():

        planilha, fr = sheet_name.split(',')
        fr = int(fr)
        # Excluir a coluna 'Danos' (se presente)
        if "Danos" in df.columns.get_level_values(0):
            df = df.drop(columns=["Danos"], level=0)

        # Capturar os nomes dos elementos do nível superior do cabeçalho
        elementos = [el for el in df.columns.get_level_values(0).unique() if el.startswith("Elemento")]

        # Processar cada elemento
        for elemento in elementos:
            # Obter as colunas Fi e Fp para o elemento
            fi_col = (elemento, 'Fi')
            fp_col = (elemento, 'Fp')

            # Calcular D para cada linha
            d_values = []
            gde_values = []
            
            for fi, fp in zip(df[fi_col], df[fp_col]):
                if fi <= 2.0:
                    d = 0.8 * fi * fp  # Equação 3.1
                elif fi >= 3.0:
                    d = (12 * fi - 28) * fp  # Equação 3.2
                else:
                    d = 0
                d_values.append(d)

            # Somatório de D para o elemento
            d_total = sum(d_values)

            # Calcular d_max para o elemento
            d_max = max(d_values)

            # Calcular gde para o elemento
            gde = d_max * (1 + ((d_total - d_max) / d_total))

            # Adicionar o valor de gde para o cálculo do gdf
            gde_values.append(gde)

            # Calcular o gde máximo para o elemento
            gde_max = max(gde_values)

            # Somatório de gde para o elemento
            gde_total = sum(gde_values)

            # Calcular o gdf para o elemento
            gdf = gde_max * sqrt(1 + gde_total - gde_max) / gde_total

            # Adicionar o resultado para este elemento com o nome da planilha
            resultados.append({
                "Planilha": planilha,
                "Elemento": elemento,
                "sum(d)": d_total,
                "d_max": d_max,
                "gde": gde,
                "gdf": fr * gdf
            })

    # Criar o DataFrame final com sum(D), d_max, gde, gdf e o nome da planilha
    result_df = pd.DataFrame(resultados)

    # Ajustar o índice para o formato final
    result_df.reset_index(drop=True, inplace=True)

    return result_df

resultado = gde('W M Pereira Junior_modelo de planilha de vistoria.xlsx')
resultado
```
